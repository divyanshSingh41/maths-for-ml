# Functions And Graphs

# Why Functions & Graphs
### In Machine Learning:

*   A model is a function
*   A loss is a function
*   Training = changing a function
*   Graphs help you see how models behave







### **Functions** - In mathematics, a function is fundamentally a **relationship between two sets of data**. It describes a specific rule: **for every input you put in, you get exactly one output**.

### Input - Output Machine
*   *x*- Input
*   *f* - Function
*   *f(x)* - Output






```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linear Function
# Example - Y = mX + C

# linear Functionn with m = 2, c = 1
def linearFunc(x):
  return 2*x + 1

l = []
for i in range(-10, 11):
  l.append(linearFunc(i))
print(l)

plt.plot(l, color="red")
plt.xlabel("x")
plt.ylabel("Y = 2x + 1")
plt.title("Graph of Y = 2x + 1")
plt.show()
```

```python
# What is Slope
# slope is a number that describes both the steepness and the direction of a line.

x = np.linspace(0, 20, 100)

# Define your lines
y1 = 3*x + 2
y2 = 4*x + 2
y3 = 5*x + 2

# Plot them
plt.plot(x, y1, label='y = 3x + 2', color='blue')
plt.plot(x, y2, label='y = 4x + 2', color='red')
plt.plot(x, y3, label='y = 5x + 2', color='green', linewidth=2)
plt.title("Slope Visualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # This creates the box identifying the lines
plt.xlim(left=0)
plt.grid(True)
# Increasing Slope

```

```python
# Negative Slope

x = np.linspace(0, 20, 100)

# Define your lines
y1 = -3*x + 2
y2 = -4*x + 2
y3 = -5*x + 2

# Plot them
plt.plot(x, y1, label='y = -3x + 2', color='blue')
plt.plot(x, y2, label='y = -4x + 2', color='red')
plt.plot(x, y3, label='y = -5x + 2', color='green', linewidth=2)

# --- THE FIX STARTS HERE ---

# 1. Get the current axes
ax = plt.gca()

# 2. Hide the top and right borders (the box)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# 3. Move the left and bottom axes to the zero position (Origin)
ax.spines['bottom'].set_position('zero') # Sets x-axis at y=0
ax.spines['left'].set_position('zero')   # Sets y-axis at x=0

# 4. Force the x-axis to show only positive numbers
plt.xlim(left=0)

# ---------------------------

plt.title("Slope Visualization (Positive Axis)")
plt.legend()
plt.grid(True)
plt.show()
```

To understand how a negative slope determines steepness, you have to separate the direction from the magnitude.                         
Here is the golden rule:
Ignore the negative sign to check steepness.

*   negative sign ($-$) just tells you the line is going downhill.
*   The number itself (absolute value) tells you how steep the drop is.

**The bigger the number (ignoring the sign), the steeper the slide.**

```python
# Quadratic Function
# Example - f(x) = x^2
def squareFunc(x):
  return x**2
l = []
for i in range(-10, 11):
  l.append(squareFunc(i))
print(l)

# graph of f(x) = x^2
plt.plot(l, color="green")
plt.xlabel("x")
plt.ylabel("f(x) = x**2")
plt.title("Graph of f(x) = x**2")
plt.show()
```

```python
# Slope Of Curve At Point
# 1. Define the data
x = np.linspace(-50, 20, 500)
y = x**3

# 2. Define the Tangent Line
# Slope at x=2 is 12(By Differentiation). Point is (2, 8).
# Equation: y - 8 = 12(x - 2)  =>  y = 12x - 16
y1 = 12 * x - 16

# 3. Plotting
plt.plot(x, y, label='y = x^3', color='black')
plt.plot(x, y1, label='Slope at x=2 (Tangent)', color='red', linestyle='--')

# 4. Adjusting Spines (The Axes)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# 5. Setting Limits (Zooming in to the relevant area)
plt.xlim(-20, 20)
plt.ylim(-3000, 3000)

plt.grid(True)
plt.legend()
plt.show()


```
