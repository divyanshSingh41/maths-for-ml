##** Indeterminate Form (0 / 0)**
If direct substitution gives 0 / 0, the limit is indeterminate.
In this case, we must simplify the fraction.

Example:
image.png

image.png

### **If the denominator becomes zero but numerator is not zero, the limit is infinite or does not exist.**
image.png

### Left-hand limit and right-hand limit must be equal

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return (x**2 - 4) / (x - 2)

x_left = np.linspace(1.5, 1.99, 200)
x_right = np.linspace(2.01, 2.5, 200)

y_left = f(x_left)
y_right = f(x_right)

plt.figure()
plt.plot(x_left, y_left)
plt.plot(x_right, y_right)
plt.axvline(2)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Limit of (x - 4)/(x - 2) as x  2")
plt.show()

# Numerical values near x = 2
for x in [1.9, 1.99, 2.01, 2.1]:
    print(f"x = {x}, f(x) = {f(x)}")
```

## **Graphical Understanding**
In the graph:
*   Blue line  values before 2
*   Orange line  values after 2
*   Vertical line at x = 2

**Even though the function is not defined at x = 2,
both sides approach y = 4.**  
**This is called a removable discontinuity.**

```python
# Other Example
def f(x):
  return (x**2 - 1)/(x - 1);

# Find limit at x = 1
x_left = np.linspace(0.5, 0.99, 200)
x_right = np.linspace(1.01, 1.5, 200)

y_left = f(x_left)
y_right = f(x_right)

plt.figure()
plt.plot(x_left, y_left, 'r')
plt.plot(x_right, y_right)
plt.axvline(1)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Limit of (x - 1)/(x - 1) as x  1")
plt.show()
```
